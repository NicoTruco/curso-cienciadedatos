## Gráfico de funciones (pyplot.plot)


Empezaremos por algo simple, creando el gráfico de una función $y = f(x)$ en el intervalo $x=[0, 1]$.

Es importante en este momento tomar en cuenta una limitación crucial que tiene la computación, el hecho de que la cantidad de información que puedo manejar es limitada *¿A qué nos referimos con esto?* A que para graficar la función para todos los valores que puede tomar $x$ entre $0$ y $1$ se necesitaría infinita información ya que hay infinitos valores posibles de $x$ en este intervalo, por lo tanto habrá infinitos valores de $y$ que corresponden a cada uno de estos.

A partir de esta limitación surge la necesidad de "[discretizar](https://es.wikipedia.org/wiki/Discretizaci%C3%B3n)" la función. En este caso, en lugar de graficar una curva perfecta para nuestra función, vamos a graficar solamente algunos puntos de ella, y los uniremos con lineas rectas. Si la cantidad de puntos que graficamos es lo suficientemente alta, la diferencia será imperceptible.

Vamos a hacer un ejemplo, crearemos cada uno de los puntos a graficar y para eso utilizaremos una lista de las coordenadas $x$ de los puntos y otra lista de la misma longitud con las coordenadas $y$ correspondientes, en este caso elegimos graficar $y = sin(2\pi.x)$ usando la función [np.sin(x)](https://docs.scipy.org/doc/numpy/reference/generated/numpy.sin.html) de la librería numpy para calcular los valores de $y$.


````python
import numpy as np

def f(x):
  return np.sin( 2 * np.pi * x )

# Cantidad de puntos
n = 10

# x/(n-1) genera valores entre 0 y 1
# ya que x va desde 0 hasta (n-1)
X = [ x/(n-1) for x in range(n) ]
print(X)

# Para cada valor en X calculo la función elegida
Y = [ f(x) for x in X ]
print(Y)
````

Usando [plt.plot(x, y)](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html) podemos generar el grafico de la función. Ya que es posible incluír muchas curvas sobre el mismo gráfico, matplotlib espera a que se termine de generar cada gráfico y al final de todo debemos llamar a [plt.show()](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.show.html) para mostrar todas las curvas generadas. La idea es en el futuro poder realizar distintos gráficos y recién al final de todo llamar una sola vez a *plt.show()* para mostrar todo junto.

*¿Cuántos puntos consideran necesarios para que el gráfico parezca una curva perfecta?*

**Tip:** Prueben cambiar la cantidad de puntos a 5, 10, 20, 50, 100, etc. para ver qué sucede.

````python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return np.sin( 2 * np.pi * x )

n = 100  # Cantidad de puntos

X = [ x/(n-1) for x in range(n) ]
Y = [ f(x) for x in X ]

plt.plot(X, Y)   # plt.plot recibe 2 listas de igual longitud
plt.show()
````

Numpy ofrece nuevos tipos de variables que permiten generar expresiones vectorizadas. De esta forma podremos crear un vector $x$ con todas las coordenadas, y aplicar la función $sin$ a todo el vector en simultáneo, lo que va a suceder es que la función se aplica a cada elemento por separado. Asimismo puedo multiplicar un vector por una constante, sumar una constante, multiplicar dos vectores, sumar dos vectores, y muchas cosas más.

Las operaciones que se realizan elemento a elemento son de mucha utilidad, pueden llegar a encontrarlas bajo el nombre *element-wise operations*.

Veamos cómo crear un vector con [np.arange(start, stop, step)](https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html) y con [np.linspace(start, stop, num)](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html):

````python
import numpy as np

t1 = np.arange(0, 1.1, 0.1)
t2 = np.linspace(0, 1, 11)

print(t1)
print(t2)
````

Con *np.arange()* debo indicar el salto entre un elemento y otro, mientras que con *np.linspace()* debo indicar la cantidad total de elementos. Además *np.arange()* no incluye el valor de *stop* en el vector mientras que *np.linspace()* sí.

Vamos a recrear el gráfico anterior usando *np.linspace()*:

````python
import matplotlib.pyplot as plt
import numpy as np

# Cantidad de puntos
n = 10

X = np.linspace(0, 1, n)
print(X)

# Se calcula la función sobre cada elemento por separado
Y = np.sin( 2 * np.pi * X )
print(Y)

plt.plot(X, Y)
plt.show()
````

Podemos ver que `2*np.pi` es una constante por lo cual al multiplicarlo por `X` se realiza la multiplicación elemento a elemento, cada elemento del vector X se multiplica por la misma constante. Luego `np.sin()` calcula el seno elemento a elemento.

## **Personalización de gráficos**

Es posible ejecutar varias veces *plt.plot()* para superponer diferentes gráficos, pero *plt.show()* se ejecuta una sola vez al final del código.

Adicionalmente, podemos personalizar cada gráfico con un tercer parámetro de la función *plt.plot()* el cual es un string. El contenido del string indicará el tipo de linea y el color. Mostramos un ejemplo donde le cambiamos el tipo de linea agregando los parámetros **m***, **r--**:

````python
import numpy as np
import matplotlib.pyplot as plt 

x = np.arange(0, 2*3.14 , 0.25)
y = np.sin(x)
plt.plot(x,y,'m1') # magenta con estrellas
plt.plot(x,np.cos(x),'rh') # rojo linea punteada
plt.show()
````

Para que todos los labels se muestren juntos, debemos ejecutar *plt.legend()* después de todos los gráficos, una sóla vez justo antes de *plt.show()*.

````python
x = np.arange(0, 2*3.14, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'b-', label='Seno') # graficamos función y1
plt.plot(x, y2, 'r-.', label='Coseno') # graficamos función y2
plt.legend() #Le decimos a maplotlib que muestre todos los labels
plt.savefig('graf3.png', dpi=300)
plt.show()
````

En caso de que no queramos mostrar las curvas superpuestas, podemos crear diferentes gráficos en la pantalla utilizando *plt.subplot(nrows, ncols, index)*.

- *nrows*: Número de filas
- *ncols*: Número de columnas
- *index*: Número entre 1 y nrows*ncols.


````python
x = np.arange(0, 2*3.14, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10,10)) #ancho, alto
plt.subplot(2, 2, 1)  # 1 fila, 2 columnas, posición 1
plt.plot(x, y1, 'b-') # graficamos función y1
plt.xlabel('eje1')
#plt.axis([0, 7, 0, 1.05])
plt.grid(linestyle='--',axis='y')
plt.title('ejemplo1')
#plt.axis([-10,10,-30,15])

plt.subplot(2, 2, 2)  # 1 fila, 2 columnas, posición 2
plt.plot(x, y2, 'r-.') # graficamos función y2
plt.xlabel('eje1.1')
plt.title('ejemplo2')

plt.subplot(2, 2, 3) 
plt.plot(x, y2, 'r-.') # graficamos función y2
plt.xlabel('eje1.1')
plt.title('ejemplo3')

plt.show()
````

Veamos algunas cosas más que pueden realizarse:
- Con `plt.figure( figsize = (w, h) )` podemos elegir el tamaño de la figura.
- Con `plt.xlabel('xlabel')` y `plt.ylabel('ylabel')` podemos ponerle nombre a los ejes.
- Con `plt.axis([xmin, xmax, ymin, ymax])` podemos elegir los límites de los ejes.
- Con `plt.title('title')` podemos mostrar un título sobre el gráfico.
- Con `plt.grid()` pueden activar una grilla cuadriculada.

Se pueden configurar muchísimos más parámetros, por lo cual les mostrarmos los que se suelen usar más regularmente, y esperemos que cuando necesiten funcionalidades más específicas puedan encontrarlas en la documentación de [matplotlib.pyplot](https://matplotlib.org/tutorials/introductory/pyplot.html).

````python
x = np.arange(0, 2*3.14, 0.3)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.title("Gráfico azul")
plt.plot(x, y1, 'bo') # graficamos función y1
plt.xlabel('Tiempo')
plt.ylabel('Corriente')

plt.subplot(1, 2, 2)
plt.title("Gráfico rojo")
plt.plot(x, y2, 'r-.') # graficamos función y2
plt.xlabel('Tiempo')
plt.axis([0, 3, -1, 1])
plt.grid()

plt.show()
````

## **Piechart**






Este grafico lo realizamos con la siguiente funcion:


> plt.**pie**($tamaños$)

````python
import matplotlib.pyplot as plt

sizes = [215, 130, 245, 210, 120, 45]
 
plt.pie(sizes)
plt.show()
````

El gráfico por sí sólo no nos dice mucho, debemos agregar información que destaque lo que queremos transmitir, estas son algunas de las opciones más útiles para este propósito:

- **explode**: Le indicamos si alguna de las porciones deben estar alejada del centro



````python
import matplotlib.pyplot as plt

sizes = [215, 130, 245, 210, 300]

e = (0.2, 0, 0, 0, 0) # Cuánto separar cada porción
 
plt.pie(sizes, explode=e)

plt.show()
````

- **labels**: Le indicamos el nombre de cada porción.


````python
import matplotlib.pyplot as plt

l = ('Python', 'C++', 'Ruby', 'Java')
sizes = [215, 130, 245, 210]

e = (0.1, 0, 0, 0)  # explode 1st slice
 
plt.pie(sizes, explode=e, labels=l)

plt.show()
````

- **colors**: Le podemos asignar un color nosotros.


````python
import matplotlib.pyplot as plt

l = ('Python', 'C++', 'Ruby', 'Java')
sizes = [215, 130, 245, 210]
c = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

e = (0.1, 0, 0, 0)  # explode 1st slice
 
plt.pie(sizes, explode=e, colors=c, labels=l)
plt.savefig('graf2.png', dpi=300)

plt.show()
````

- **startangle**: Le indicamos el angulo de rotación.


````python
import matplotlib.pyplot as plt

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [2150, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

explode = (0.1, 0, 0, 0)  # explode 1st slice
 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,  startangle=160)

plt.show()
````

- **autopct**: Le indicamos que muestre los porcentajes. Este parámetro sigue una sintáxis estándar, para que muestre 1 decimal se usa: "%1.1f%%". Esto le dice que vamos a mostrar un numero float con 1 decimal y el %% indica que queremos el simbolo %. 

````python
import matplotlib.pyplot as plt

labels = ('Python', 'C++', 'Ruby', 'Java')
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

explode = (0.1, 0, 0, 0)  # explode 1st slice
 
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%5.01f%%',  startangle=140)
 #en autopct ponemos como deseamos que se escriban los porcentajes

plt.show()
````

## **Histograma**

Para realizar un histograma utilizamos la función:

> plt.**hist**(x, bins, propiedades)

- x: secuencia de datos
- bins: número de separaciones de igual ancho

Veamos un ejemplo:

````python
from matplotlib import pyplot
import matplotlib.pyplot as plt
import numpy as np

rand = np.random.normal(0,2,1000) # generamos 1000 numeros al azar con distribución normal
#print(rand)
plt.hist(rand,100,density=True)
plt.show()
````

Podemos superponer la función a la que tiende la distribución (a modo de ejemplo, no es necesario entender el procedimiento matemático):

**Nota**: Con el parametro **densit=True** indicamos que normalize el histograma, esto hará que la suma de todas las columnas sea igual a 1.

````python
import matplotlib.pyplot as plt
import numpy as np
mu = 0
var = 1
t = np.arange(-10,10,0.01)
y = 1/np.sqrt(2*np.pi*var)*np.exp(-(t-mu)**2/(2*var))
rand = np.random.normal(mu,var,10000)
plt.hist(rand,100,density=True)
plt.plot(t,y)
plt.savefig('graf2.png', dpi=300)
plt.show()
````

## **Stem plot**
Un stem plot nos permite graficar puntos, se utiliza de la siguiente manera:


> plt.**stem**(x,y)




````python
import matplotlib.pyplot as plt
import numpy as np

puntos = 10
espacio = 1

inicio = 0 
fin =  5

x = np.linspace(inicio,fin, puntos)
y = x**2 - 4
y2 = -0.5*x**2-4

plt.subplot(2, 1, 1)
plt.stem(x,y)
plt.subplot(2, 1, 2)
plt.stem(x,y2, 'r-')

plt.show()
````

Al igual que los otros tipos de gráficos, podemos personalizar parámetros como el formato de linea (*linefmt*), el formato del punto (*markerfmt*), el formato de la base (*basefmt*) y el nombre:

**Nota**: Podemos combinar este grafico con un plot común!

````python
import matplotlib.pyplot as plt
import numpy as np

puntos = 10
espacio = 1

inicio = 0 
fin =  5

x = np.linspace(inicio,fin, puntos)
y = x**2-4
y2 = -0.5 * x**2 - 4

plt.stem(x, y , linefmt='g--', markerfmt='go', basefmt='k', label='Ingresos')
plt.stem(x, y2, linefmt='r-.', markerfmt='ro', basefmt='k', label = 'Gastos')
plt.plot(x,y+y2,'b-',label = 'Ganancia')

plt.legend()
plt.savefig('graf1.png', dpi=300)

plt.show()
````

$$f(x,y) = cos(x) + sin^2(y)$$

````python
def f(x, y):
    return np.cos(x) + np.sin(y) ** 2
````

````python
x = np.linspace(-2, 2)
y = np.linspace(-2, 2)
xx, yy = np.meshgrid(x, y)

plt.contour(xx, yy, f(xx, yy))
plt.colorbar()
````

> La función *contourf* es casi idéntica pero rellena el espacio entre niveles. 


> - Podemos especificar manualmente estos niveles usando el cuarto argumento:

````python
zz = f(xx, yy)
plt.contourf(xx, yy, zz, np.linspace(-0.5, 2.0),cmap=plt.cm.rainbow)
plt.colorbar()
````

### Multiples ejes

A veces es útil tener varios ejes y en el mismo gráfico.

Por ejemplo, si queremos investigar la correlación entre la tasa de obesidad de un país y la cantidad de calorías que ingieren las personas en este país por cada comida. 

````python
# Creando datos
calories = [380.70, 420.98, 454.91, 406.45, 446.16, 498.08, 504.54, 459.05, 459.55, 484.79]
countries = ['India', 'Japan', 'Korea', 'China', 'Thai', 'Italy', 'France', 'Greece', 'Mexico', 'US']
obesity_rates = [3.9, 4.3, 4.7, 6.2, 10, 19.9, 21.6, 24.9, 28.9, 36.2]

# Plot
fig, ax1 = plt.subplots(figsize = (10, 8))
ax1.bar(countries, obesity_rates, color='C8')
ax1.set_ylabel('obesity rate(%)', color='C8')
ax1.tick_params(axis='y', labelcolor='C8')

# Permite multiples ejes
ax2 = ax1.twinx()
ax2.plot(countries, calories, color='C0')
ax2.set_ylabel('calories', color='C0')
ax2.tick_params(axis='y', labelcolor='C0')
plt.show()

````

### 3D Faces
Besides of contour plot, we can directly plot the 3D faces as well using `matplotlib`.

Plot the 3D-face of 2D [joint Gaussian distribution](https://en.wikipedia.org/wiki/Multivariate_normal_distribution).

$\boldsymbol{\mu}  = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $, $\boldsymbol{\Sigma}  = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} $.

> Note: $\mu$ is location of center. and $\Sigma$ is covariance matrix

````python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

Mu = np.array([0, 0])
Cov = np.array([[1, 0], [0, 1]])
rv = multivariate_normal(Mu, Cov)

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.25)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
Z = rv.pdf(pos)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet)
plt.show()
````

Cambiamos la ubicación y la matriz de covarianza

$\boldsymbol{\mu}  = \begin{bmatrix} 0 \\ 4 \end{bmatrix} $, $\boldsymbol{\Sigma}  = \begin{bmatrix} 5 & 0 \\ 0 & 1 \end{bmatrix} $.

````python
Mu = np.array([0, 4])
Cov = np.array([[5, 0], [0, 1]])
rv = multivariate_normal(Mu, Cov)

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.25)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
Z = rv.pdf(pos)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet)
plt.show()
````

$\boldsymbol{\mu}  = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $, $\boldsymbol{\Sigma}  = \begin{bmatrix} 10.5 & -9.5 \\ -9.5 & 10.5 \end{bmatrix} $.

````python
Mu = np.array([0, 0])
Cov = np.array([[10.5, -9.5], [-9.5, 10.5]])
rv = multivariate_normal(Mu, Cov)

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.25)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
Z = rv.pdf(pos)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet)
plt.show()
````
