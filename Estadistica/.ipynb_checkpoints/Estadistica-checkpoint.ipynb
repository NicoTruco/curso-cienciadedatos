## Medidas de tendencia central
### Media, Mediana y Moda

```python
import matplotlib.pyplot as plt
pd.options.display.max_columns=None#permite visualizar todas las columnas
```

```python
df.head()
```

```python
#viendo el precio en dolares
df["price_usd"]
```

```python
#promedio
df["price_usd"].mean()
```

```python
#nos interesa analizar los atipicos, es decir cual es el auto con mayor precio
#cual de todos es el mas barato
```

```python
#mediana
df["price_usd"].median()
```

A partir de la media y a mediana podemos observar que la mediana es mucho menor (4800) que la media(6639) es decir existen autos muy costosos que aumentan la media y la alejan de la mediana, en el caso de una distribucion normal la media==mediana

Realizaremos ahora un histograma(generalizacion de grafico de barras por intervalos) de los precios

```python
df["price_usd"].plot.hist(bins=20, edgecolor="black")#bins son los intervalos de valores#
plt.show()
```

# Medidas de dispersion

```python
#desviacion estandar
df["price_usd"].std()
```

```python
#Rango=valormax-valormin
rango=df["price_usd"].max()-df["price_usd"].min()
print(rango)
```

```python
#Calcular los cuartiles
#Recordar que la mediana es el cuartil 2 (Q2)
median=df["price_usd"].median()
Q1=df["price_usd"].quantile(q=0.25)
Q3=df["price_usd"].quantile(q=0.75)

#calculemos el valor maximo y minimo con cuartiles

min_val=df["price_usd"].quantile(q=0)
max_val=df["price_usd"].quantile(q=1)
print(min_val,Q1,median,Q3,max_val)
```

#### Conclusiones
Realizamos un dataframe tomando como marca a Nissan y modelo a Qashqai. Luego procedemos a graficar los precios de los automoviles considerando su año de producción.

1.-De esto podemos observar que los precios de estos autos en su mayoria son medios (entre 7000 y 15000)

2.-Los autos mas caros corresponden a los mas nuevos

3.-Los autos mas nuevos tiene un salto en precio y se separan del resto.

```python
#calculo de rango intercuartil
iqr=Q3-Q1
iqr
```

```python
minlimit=Q1-1.5*iqr
maxlimit=Q3+1.5*iqr
print(minlimit,maxlimit)
```
