```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("iris.csv",index_col=0)
```

```python
data
```

## loc y Iloc
-loc con etiqueta
-iloc con posicion

```python
plt.hist(data.iloc[:,0],edgecolor="black")
plt.title("Distribución del largo del Sepalo en cm")
plt.show()
```

```python
data.iloc[:,0].mean()#promedio del largo del sepalo
```

```python
data["SepalLengthCm"].mean()
```

```python
plt.hist(data.iloc[:,1],edgecolor="black")
plt.title("distribución del ancho del sepalo en cm")
plt.show()
```

```python
#la media o promedio
data.iloc[:,1].mean()
```

```python
plt.hist(data.iloc[:,2],edgecolor="black")
plt.title("distribución del largo del petalo en cm")
plt.show()
```

```python
data.iloc[:,2].mean()
```

```python
plt.hist(data.iloc[:,3],edgecolor="black")
plt.title("distribución del ancho del petalo en cm")
plt.show()
```

```python
data.iloc[:,3].mean()
```

```python
#distribuciones
plt.scatter(data["PetalLengthCm"],data["PetalWidthCm"])
plt.title("Dimensiones de los petalos")
plt.show()
```

```python
index=data[data["PetalLengthCm"]>=2.5].index
index
```

```python
data.loc[index,"Tamaño"]="Grande"
data.loc[index,"Color"]="r"
data
```

```python
#caso contrario
indices=data[data["PetalLengthCm"]<2.5].index
data.loc[indices,"Tamaño"]="Pequeña"
data.loc[indices,"Color"]="b"
data
```

```python
#incluyendo el color en el scatter
plt.scatter(data["PetalLengthCm"],data["PetalWidthCm"],c=data.loc[:,"Color"])
```

```python
data[data["PetalLengthCm"]<2.5]
```
