```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#usaremos un dataset que viene por defecto en seaborn#
iris=sns.load_dataset("iris")
iris.head()
```

```python
#realizaremos un scatterplot#

sns.scatterplot(data=iris, x="sepal_length",y="petal_length",hue="species")
plt.grid()
plt.show()
```

```python
#joint plot
sns.jointplot(data=iris,x="sepal_length",y="petal_length",hue="species")
plt.show()
```

```python
#boxplot
sns.boxplot(data=iris,y="sepal_length",x="species")
plt.show()
```

```python
#barplot
sns.barplot(data=iris,x="species",y="sepal_length")
plt.show()
```

```python
sns.jointplot(data=iris,y="petal_length",x="sepal_length",hue=("species"=="setosa"))
plt.show()
#revisar!?#
```
