### Introducción

Starcraft corresponde a un franquicia de videojuegos de estrategia en tiempo real. Sus acontecimientos tienen lugar en un sector de la galaxia llamado sector Koprulu. En este juego existen tres razas principales Terran, Zerg y Protoss las cuales difieren en historia y estilo de juego.Este titulo tuvo un auge durante finales de los 90's y una gran participación a niveles competivos. Es por ello que fue escogida para realizar un estudio estadístico de variables tomadas durante una liga de prueba. En este caso el dataset escogido corresponde a datos de jugadores de Starcraft II .Los datos corresponden a los siguientes:

1.- GameID: número unico de cada jugador para cada juego.

2.- LeagueIndex: números ordinales que denominan la medalla de cada jugador del 1 al 8 en el siguiente orden: Bronze, Silver, Gold, Platinum, Diamond, Master, GrandMaster y Professional.

3.-Age: edad

4.- HourPerWeek: Horas reportadas gastadas jugando por semana.

5.- TotalHours: Horas totales gastadas jugando.

6.- APM: acciones por minuto.

7.-SelectByHotkeys: Número de unidades o edificios seleccionadas apartir de hotkeys por periodo de tiempo.

8.- AssigntoHotkeys: Número de unidades o edificios asignadas a hotkeys en un periodo de tiempo.

9.- UniqueHotkeys: Número de hotkeys unias usadas por periodo de tiempo.

10.- MinimapAttacks: Número de acciones de ataque en el minimapa por periodo de tiempo.

11.- MinimapRightClicks: Número de clicks derechos en el minimapa por periodo de tiempo.

12.-NumberOfPACS: Número de PACS en un periodo de tiempo. Los PACS corresponden a movimientos de cámara donde se realiza alguna acción y luego se vuelve a mover la camara. ref https://www.youtube.com/watch?v=LeZvZzyfJ0E

13.- GapBetweenPACS: duración media en milisegundos entre PACS.

14.- ActionLatency: Latencia media desde el inicio de un PAC hasta que suceda la primera acción en milisegundos.

15.- ActionsInPac: Número medio de actiones entre cada PAC.

16.- TotalMapExplored: Número de coordenadas en una grilla de 24x24 vistas por el jugador en un periodo de tiempo.

17.-WorkersMade: Numero de SCVs, drones y probes producidos en un periodo de tiempo.

18.- UniqueUnitsMade: Unidades unicas producidas en el periodo de prueba.

19.- ComplexUnitsMade: Número de ghosts,infestors y high templars entrenados en el periodo de prueba.

20.- ComplexAbilitiesUsed: Habiliides que requieren alguna instrucción con algun target especifico usadas en el periodo de prueba.

Ahora procederemos a calcular las variables de dispersión para continuar el análisis.

```python
#Cálculo de desviaciones estandar
std_apm=df["APM"].std()
std_map=df["TotalMapExplored"].std()
print("std_apm={0}     std_map={1}".format(std_apm,std_map))
```

```python
#Cálculo de medianas
median_apm=df["APM"].median()
median_map=df["TotalMapExplored"].median()
print("median_apm={0}     median_map={1}".format(median_apm,median_map))
```

```python
#Cálculo de minimos y maximos
min_apm=df["APM"].min()
max_apm=df["APM"].max()
min_map=df["TotalMapExplored"].min()
max_map=df["TotalMapExplored"].max()
print("min_apm={0}/max_apm={1}".format(min_apm,max_apm))
print("min_map={0}/max_map={1}".format(min_map,max_map))
```

```python
#Rangos
#Utilizaremos los maximos y minimos para cálcular el rango de las variables de interés
rango_apm=max_apm-min_apm
rango_map=max_map-min_map
print("Rango APM={0}/Rango MAP={1}".format(rango_apm,rango_map))
```

De lo anterios podemos concluir que la diferencia de acciones por minuto entre el jugador que realiza menos acciones) y el que realiza mas es de 367 acciones. Asimismo en el caso de la exploración el jugador que mas exploro el mapa lo hizo en 53 cuadrillas de diferencia con el que lo hizo menos.

```python
#Cálculo de cuartiles
Q1_apm=df["APM"].quantile(q=0.25)
Q3_apm=df["APM"].quantile(q=0.75)
Q1_map=df["TotalMapExplored"].quantile(q=0.25)
Q3_map=df["TotalMapExplored"].quantile(q=0.75)
#Mapeo general de las variables
#APM
print("APM:",min_apm,Q1_apm,median_apm,Q3_apm,max_apm)
#TotalMapExplored
print("TotalMapExplored:",min_map,Q1_map,median_map,Q3_map,max_map)
```

De lo expuesto anteriormente podemos notar lo lejos que se encuentran los valores máximos para ambos casos en comparación con el cuartil 3. Es por ello que tenemos como hipotesis que estos valores muy altos corresponderan a outliers en las distribuciones que estamos estudiando. Esta hipotesis la comprobaremos a través del rango intercuartil y de los gráficos que presentaremos a continuación.

```python
#Cálculo de rango intercuartil
iqr_apm=Q3_apm-Q1_apm
iqr_map=Q3_map-Q1_map
print("iqr_apm=",iqr_apm," iqr_map=",iqr_map)
```

```python
#cálculo de limite superior e inferior para distribuciones gaussianas
#Para comprobar la hipotesis haremos el cálculo de los limites como sigue:
minlimit_apm=Q1_apm-1.5*iqr_apm
maxlimit_apm=Q3_apm+1.5*iqr_apm
minlimit_map=Q1_map-1.5*iqr_map
maxlimit_map=Q3_map+1.5*iqr_map
print("APM= min:{0}/max:{1}".format(minlimit_apm,maxlimit_apm))
print("MAP= min:{0}/max:{1}".format(minlimit_map,maxlimit_map))
```

Observamos que en el caso de las APM el limite inferior nos da fuera de rango por lo que notamos que no es un metodo adecuado para verificar los puntos outliers. Sin embargo como primera idea nos muestra que en el limite superior de ambas variables nuestra hipotesis podría ser correcta donde los valores máximos comentados anteriormente superan los limites superiores mostrados en este cálculo. Para corroborarlo realizaremos gráficos de caja y veremos los resultados.

### Normalización Z-score

En este punto del análisis nos piden realizar una normalización y luego comparar distintos tipos de escalamiento por lo que usaremos la función timeit para compararlas.

```python
import timeit
from sklearn import datasets, linear_model
#importamos datasets y un modelo de regresion lineal#
from sklearn.preprocessing import MinMaxScaler
```

```python
#Preparamos los datos que utilizaremos
scaler=MinMaxScaler()
raw_sbh_first=df["SelectByHotkeys"]
raw_al_first=df["ActionLatency"]
y=df["APM"]
raw_sbh_2=raw_sbh_first.to_numpy()
raw_al_2=raw_al_first.to_numpy()
raw_sbh=raw_sbh_2[:,None]
raw_al=raw_al_2[:,None]
```

```python
#Escalamiento max-min
max_raw_sbh=max(raw_sbh)
min_raw_sbh=min(raw_sbh)
max_raw_al=max(raw_al)
min_raw_al=min(raw_al)
scaled_sbh=(2*raw_sbh-min_raw_sbh)/(max_raw_sbh-min_raw_sbh)
scaled_al=(2*raw_al-min_raw_al)/(max_raw_al-min_raw_al)
```

```python
#Z-score
promedio_sbh=raw_sbh.mean()
promedio_al=raw_al.mean()
std_sbh=raw_sbh.std()
std_al=raw_al.std()
scaled_z_sbh=(raw_sbh-promedio_sbh)/std_sbh
scaled_z_al=(raw_al-promedio_al)/std_al
```

```python
#definiendo entrenamientos para comparar tiempos
def train_raw_sbh():
    linear_model.LinearRegression().fit(raw_sbh,y)
    
def train_raw_al():
    linear_model.LinearRegression().fit(raw_al,y)
    
def train_scaled_sbh():
    linear_model.LinearRegression().fit(scaled_sbh,y)
    
def train_scaled_al():
    linear_model.LinearRegression().fit(scaled_al,y)
    
def train_scaled_z_sbh():
    linear_model.LinearRegression().fit(scaled_z_sbh,y)
    
def train_scaled_z_al():
    linear_model.LinearRegression().fit(scaled_z_al,y)
```

```python
#Midiendo tiempos
raw_time_sbh=timeit.timeit(train_raw_sbh,number=100)
raw_time_al=timeit.timeit(train_raw_al,number=100)
scaled_time_sbh=timeit.timeit(train_scaled_sbh,number=100)
scaled_time_al=timeit.timeit(train_scaled_al,number=100)
scaled_time_z_sbh=timeit.timeit(train_scaled_z_sbh,number=100)
scaled_time_z_al=timeit.timeit(train_scaled_z_al,number=100)
```

```python
#Resultados de tiempos
print("Selection by hotkeys")
print("train raw:{}".format(raw_time_sbh))
print("train scaled:{}".format(scaled_time_sbh))
print("train scaled z:{}".format(scaled_time_z_sbh))
print("====================================================")
print("Action Latency")
print("train raw:{}".format(raw_time_al))
print("train scaled:{}".format(scaled_time_al))
print("train scaled z:{}".format(scaled_time_z_al))
```

Ahora procedemos a gráficas todos los conjutos de datos obtenidos anteriormente.

```python
fig, axs = plt.subplots(3,1,sharex=True)
axs[0].hist(raw_sbh)
axs[1].hist(scaled_sbh)
axs[2].hist(scaled_z_sbh)
plt.show()
```

```python
fig, axs = plt.subplots(3,1,sharex=True)
axs[0].hist(raw_al)
axs[1].hist(scaled_al)
axs[2].hist(scaled_z_al)
plt.show()
```

Continuando con nuestro análisis realizaremos un gráfico "pairplot" para observar de antemano como debiese comportarse nuestra matriz de covarianza

```python
sns.pairplot(penguins_clean,height=5,aspect=.8,kind="reg")
plt.show()
```

```python
#Desarrollo de matriz de correlación
scaler=StandardScaler()
scaled=scaler.fit_transform(penguins_clean[["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]])
covariance_matrix = np.cov(scaled.T)
#grafico
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm=sns.heatmap(covariance_matrix,
              cbar=True,#escala de colores
              annot=True,
              square=True, #cuadrado perfecto
              annot_kws={"size":12},
              yticklabels=["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"],
              xticklabels=["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"])
```

Del gráfico anterior podemos observar que nuestra matriz de covarianza se comporta igual que el gráfico "pairplot" presentado anteriormente, es decir, muestra las pendientes de las regresiones lineales de cada uno que nos indican si son correlaciones positivas, negativas o si no existe correlacion.

En este caso en particular tenemos que nuestras variables entre si se comportan con correlaciones positivas y negativas sin existir no correlación entre ellas, es decir, cada una de las variables esta relacionada con cada una de las otras.
