```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("medusas.csv",sep=";")
```

### Data

En este caso hemos escogido datos reales de dos fuentes distintas de información. La primera de ellas corresponde al consumo de helado en españa en el 2020 por mes en miles de kilos.(https://www.statista.com/statistics/765072/consumption-from-ice-creams-from-the-homes-spanish-people-month-to-month/)

En segundo conjunto de datos corresponde a las picaduras de medusa en Santa Pola (ciudad española) durante 2020 medidos en (picaduras/km de playa/dia)(https://www.santapola.es/wp-content/uploads/bsk-pdf-manager/2021/03/Informe-Medusas-Santa-Pola-2020.pdf).

Ambas variables no son causales ninguna de la otra pero si tienen un fenomeno en común que corresponde las altas temperaturas.

En este caso poseemos poca cantidad de data dado que queriamos utilizar datos verdaderos para demostrar que es posible que exista correlación pero sin existir causalidad.

```python
df
```

```python
sns.jointplot(data=df,x="Volumen de helado 2020 en miles de kilos",y="Picaduras de medusa 2020 en Santa Pola(picaduras/km/dia)",kind="reg")
plt.show()
```

En los datos anteriores rellenamos los datos que no corresponden al periodo junio-septiembre del conjunto de datos de picaduras dado que en el estudio solo consideraba esos meses con picaduras. Aún cuando rellenamos los datos con los supuestos y que ambos datos no se relacionan directamente entre sí podemos observar que la recta que mejor se aproxima a los datos propuestos tienen una pendiente posivita por lo que creemos que puede tener una correlación.

```python
sns.heatmap(df.corr(),annot=True)
```

Observamos como la correlación entre ambos da 0.7 que corresponde a un numero muy cercano a 1 por lo que podemos decir que ambos conjuntos de datos se correlacionan.

Aqui tenemos un claro ejemplo de como dos variables que no tienen que ver entre si pero que responden a un fenomeno en común se correlacionan y podriamos caer en el error de pensar de son causales alguna de ellas con la otra.
