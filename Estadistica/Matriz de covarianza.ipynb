```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

iris=sns.load_dataset("iris")
```

```python
sns.pairplot(iris,height=5,aspect=.8,kind="reg")
plt.show()
```

```python
#normalizacion de los datos
scaler = StandardScaler()
scaled = scaler.fit_transform(iris[['sepal_length','sepal_width','petal_length','petal_width']])
```

```python
#Standard Scaler es una normalización que tiene en cuneta restar por el promedio y dividir por la desviación estandar.
scaled.shape
```

```python
#para la matriz de covarianza necesitamos transponer
#el scaled
scaled.T.shape
```

```python
#realizamos la matriz de covarianza
covariance_matrix=np.cov(scaled.T)
covariance_matrix
```

```python
# la diagonal corresponde a los valores que estan relacionado con ellas mismas
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm=sns.heatmap(covariance_matrix,
              cbar=True,#escala de colores
              annot=True,
              square=True, #cuadrado perfecto
              annot_kws={"size":12},
              yticklabels=['sepal_length','sepal_width','petal_length','petal_width'],
              xticklabels=['sepal_length','sepal_width','petal_length','petal_width'])
```

<h5 style="text-align:center"> Covarianza </h5>

-Cuando entre las dos variables hay una <span style="font-weight:bold;color=green">relación directa</span>, la covarianza da un valor positivo.

-Cuando entre las dos variables hay una relación inversa, la covarianza da un valor negativo.

-Cuando entre las dos variables no hay relación, la covarianza da un valor entorno a cero.

```python
sns.heatmap(iris.corr(),annot=True)#mismo gráfico pero mas resumido#
#annot permite poner numeros en cada coordenada#
```
