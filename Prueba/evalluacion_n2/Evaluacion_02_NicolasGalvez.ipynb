```python
import numpy as np
import pandas as pd
datos = {'X': [1, 2, 3, 4, 1], 'Y': [5, 2, 3, 2, 0], 'Z': [2, 2, 5, 7, 1]}
df = pd.DataFrame(data=datos)
print(df)
print("-------------")
#eliminamos la columna Y para dejar las otras(dataframe de dos columnas)#
columnasXZ=df.drop(["Y"],axis=1)
print(columnasXZ)
print("-------------")
#agregamos nueva columna(agregar una columna)#
columnasXZ=columnasXZ.assign(W=[21, 32, 43, 54, 61])
print(columnasXZ)
```

# Ejercicio



Leer el archivo employees.csv

Mostrar los primero 10 registros

Obtener los registros donde el nombre de un empleado empiece con *J*.


Obtener aquellos empleados donde su apellido no contenga la letra *h*.

Ordenar en modo descendente los empleados a partir de su ID de departamento.

Obtener el empleado que es gerente de la empresa.

Encontrar todos los empleados que están a cargo de un jefe.

Encontrar los empleados cuyo nombre termina en la letra a.

Obtener empleados cuyo nombre empiece ya sea por _a_, _l_ o _r_. Ordenar por ID de departamento.

Obtener los empleados que pertenezcan ya sea a los departamentos con ID 80 ó 30.

Obtener empleados cuyo jefe tenga como ID 100, 124, o 149.

Encontrar aquellos empleados que en su nombre la tercera letra es una _r_.

Obtener los empleados que **no** pertenezcan a los departamentos con ID 80 ó 30.

```python
import re
muestra=pd.read_csv("employees.csv",sep=",")
#imprimir los 10 primeros registros
print("10 primeros registros")
print(muestra.head(10))
```

```python
#imprimir los registros donde el nombre comienza con J"
nombres_con_jota=muestra["first_name"].str.contains("^J",regex=True)
muestra_j=muestra.assign(empiezaconJ=nombres_con_jota)
print("-----------------------------------------------------------------------------------------------")
print("registros cuyo nombre empiezan con J")
nombres_emp_j=muestra_j[muestra_j["empiezaconJ"]==True]
print(nombres_emp_j)
```

```python
#imprimir los registros que no contengan h en su apellido"
import re
apellidos_con_h=muestra["last_name"].str.contains("h|H",regex=True)
muestra_h=muestra.assign(tieneh=apellidos_con_h)
print("-----------------------------------------------------------------------------------------------")
print("registros cuyo apellido no tienen h")
apellidos_sin_h=muestra_h[muestra_h["tieneh"]==False]
print(apellidos_sin_h)
```

```python
#obtener lista ordenada decendentemene segun employ_id#
descendiente = muestra.sort_values('employ_id',ascending=False)
print(descendiente)
```

```python
#Obtener al gerente de la empresa#
#primero veremos los cargos existentes en el listado#
print(muestra["job_id"].unique())
print("-------------------------------------------------")
#los gerentes seran los que tengan la sigla MGR en su nomnbre es decir FI_MGR y AC_MGR#
gerentes=muestra[(muestra["job_id"]=="AC_MGR") | (muestra["job_id"]=="FI_MGR")]
print(gerentes)
```

```python
#obtener empleados que tienen un jefe#
#son aquellos que tienen un numero en manager id#
#borraremos aquenos que no tengan dato en manager id#
muestra.dropna(subset=['manager_id'])
```

```python
#Encontrar los empleados cuyo nombre termina en la letra a#
nombres_fin_a=muestra["first_name"].str.contains("a$",regex=True)
muestra_a=muestra.assign(Terminacona=nombres_fin_a)
print("-----------------------------------------------------------------------------------------------")
print("registros cuyo nombre termina con a")
nombres_fin_a=muestra_a[muestra_a["Terminacona"]==True]
print(nombres_fin_a)
```

```python
#Obtener empleados cuyo nombre empiece ya sea por a, l o r. Ordenar por ID de departamento.
nombres_con_alr=muestra["first_name"].str.contains("^A|^L|^R",regex=True)
muestra_alr=muestra.assign(empiezaconalr=nombres_con_alr)
print("-----------------------------------------------------------------------------------------------")
print("registros cuyo nombre empiezan con A L o R")
nombres_emp_alr=muestra_alr[muestra_alr["empiezaconalr"]==True]
nombres_emp_alr.sort_values("department_id")
```

```python
#Obtener los empleados que pertenezcan ya sea a los departamentos con ID 80 ó 30
print(muestra["department_id"].unique())
print("-------------------------------------------------")
departamentos=muestra[(muestra["department_id"]==30) | (muestra["department_id"]==80)]
print(departamentos)
```

```python
#Obtener empleados cuyo jefe tenga como ID 100, 124, o 149.
jefes=muestra[(muestra["manager_id"]==100) | (muestra["manager_id"]==124)|(muestra["manager_id"]==149)]
jefes=jefes.sort_values("manager_id")
print(jefes)
```

```python
#Encontrar aquellos empleados que en su nombre la tercera letra es una r.
nombres_con_3r=muestra["first_name"].str.contains("^..r",regex=True)
muestra_3r=muestra.assign(tienetercerar=nombres_con_3r)
print("-----------------------------------------------------------------------------------------------")
print("registros cuyo nombre tiene una r en la tercera posicion")
nombres_con_3r=muestra_3r[muestra_3r["tienetercerar"]==True]
print(nombres_con_3r)
```

```python
#Obtener los empleados que no pertenezcan a los departamentos con ID 80 ó 30.
no_departamento=muestra[(muestra["department_id"]!=80) & (muestra["department_id"]!=30)]
print(no_departamento["department_id"].unique())
print(no_departamento)
```
